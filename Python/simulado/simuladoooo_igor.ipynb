```python
a = int(input("Digite o lado A do triângulo: "))
b = int(input("Digite o lado B do triângulo: "))
c = int(input("Digite o lado C do triângulo: "))

triangulo = True

if abs(b-c)<a<b+c:
    triangulo = False
elif abs(a-c)<b<a+c:
    triangulo = False
elif abs(a-b)<c<a+b:
    triangulo = False
if triangulo == False:
    print("não é triângulo.")

if triangulo == True:
    if a == b and b == c:
        print("triângulo equilátero!")
    elif a != b and b != c and c != a:
        print("triângulo escaleno!")
    else:
        print("triângulo isósceles!")
```

```python
import random as rd

aleatorio = rd.randint(1,1000)
tentativa = 0

while True:
    palpite = int(input("Dê seu palpite: "))
    tentativa += 1

    if (palpite - aleatorio) != 0 and abs(palpite - aleatorio) <= 3:
        print("Quase!!")
    elif palpite > aleatorio:
        print("Muito alto...")
    elif palpite < aleatorio:
        print("Muito baixo...")
    elif palpite == aleatorio:
        break
    else:
        print("só números!")

print(f"Voce acertou! O número é {aleatorio}.\nForam necessárias {tentativa} tentativas.")


```

```python
alunos = []

while True:
    nome = input("Digite o nome do aluno (ou 'sair' para encerrar): ")
    if nome == 'sair':
        break
    idade = int(input("Digite a idade do aluno: "))

    mat = int(input("Nota em matemática: "))
    port = int(input("Nota em português: "))
    cie = int(input("Nota em ciências: "))

    notas = (mat,port,cie)
    media = round((sum(notas)/3),2)
    aluno = {'nome':nome,'idade':idade,'media':media}

    alunos.append(aluno)

print("Relatório dos alunos cadastrados: ")

for i in alunos:
    print(f"Nome: {i['nome']} - Média: {i['media']}")

```
