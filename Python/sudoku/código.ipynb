```python
import numpy as np

def printar_jogo():
    
    with open('template.txt','r') as arq:
        
        for i in arq.readlines():
            linha = []
            for j in i:
                if j.isnumeric():
                    print(j, end=" ")
                    linha.append(j)
            print()
            matriz.append(linha)

matriz = []
printar_jogo()

def verifica_jogada(x,y,valor):
    
    array = np.array(matriz)
    
    validacao = True

    #valida se está vazio
    if matriz[x][y] == 0:
        validacao = False
    
    #validação por linha
    for i in matriz[x]:
        if valor == int(i):
            validacao = False
            break
    
    #validação por coluna
    for i in matriz:
        if valor == int(i[y]):
            validacao = False
            break
    
    #descobrindo quais quadrantes se encaixam na linha
    if x < 3:
        a = 0
    elif x < 6:
        a = 3
    elif x < 9:
        a = 6

    #descobrindo quais quadrantes se encaixam na coluna
    if y < 3:
        b = 0
    elif y < 6:
        b = 3 
    elif y < 9:
        b = 6
    
    #validação por quadrante
    quadrante = array[a:a+3,b:b+3]
    for i in quadrante:
        for j in i:
            if valor == int(j):
                validacao = False
                
    return validacao

linha = int(input("Digite a linha que deseja jogar: "))
coluna = int(input("Digite a coluna que deseja jogar: "))
valor = int(input("Digite o valor que deseja inserir: "))

verifica_jogada(linha,coluna,valor)
    
```
