exercicio 1 - desenvolva um código que recebe um ano e valida se ele é ou não bissexto, sabendo que além de precisar ser divisível exato de 4 (o resto da divisão do ano por 4 precisa ser 0), a partir do ano 0, e a cada 100 anos não é um ano bissexto a menos que seja múltiplo de 400.

```python
ano = int(input("Digite um ano: "))

if ano < 1:
    print("Ano inválido!")

elif ano % 4 == 0 and not ano % 100 == 0:
    print("Ano bissexto.")

elif ano % 400 == 0:
    print("Ano bissexto.")

else:
    print("Não é bissexto.")
```

exercício 2 - sorteie um número entre um intervalo desejado para ser seu número secreto. o usuário deve inserir o ponto "a" e o ponto "b", e deve ser validado se o número secreto está entre os pontos "a" e "b", se ele está acima dos valores "a" e "b", se está abaixo ou se é um dos dois. após isso deve ser dada a oportunidade do usuário tentar adivinhar o número secreto, retorne a ele se acertou ou não.

```python
import random as r

sorteio = r.randint(1,100)

while True:
    a = int(input("Digite o valor de A: (1 a 100)"))
    b = int(input("Digite o valor de B: (1 a 100)"))

    if a < 1 or b < 1 or a > 100 or b > 100:
        print("O valor deve ser entre 1 e 100!")
    else:
        break

if sorteio > a and sorteio < b:
    print("Você acertou! O número secreto está entre o intervalo digitado.")

elif sorteio > b:
    print("O número secreto está acima do intervalo digitado.")

elif sorteio < a:
    print("O número secreto está abaixo do intervalo digitado.")

elif sorteio == a or sorteio == b:
    print("Você acertou o número através de um dos pontos!")

print(f"Número sorteado: {sorteio}")
```

exercício 3 - o usuário deve inserir os tamanhos dos lados "a", "b" e "c", e receber a validação se o triângulo formado por estes lados é: equilátero, isósceles ou escaleno.

```python
while True:

    a = int(input("Digite o lado A do triângulo: "))
    b = int(input("Digite o lado B do triângulo: "))
    c = int(input("Digite o lado C do triângulo: "))

    if a < 1 or b < 1 or c < 1:
        print("Erro!")
    else:
        break

if a == b and b == c:
    print("Triângulo equilátero!")

elif a != b and b != c and c != a:
    print("Triângulo escaleno!")

else:
    print("Triângulo isósceles!")
```

Faça um programa que imprime uma pirâmide formatada no terminal a partir de uma base fornecida pelo usuário
exemplo: base = 5

output:

    X
   X X
  X X X 
 X X X X
X X X X X

```python
base = int(input("Digite a base da pirâmide: "))

piramide = []

for i in range(base):
    piramide.append("X")
    print(' ' * (base - i - 1), " ".join(piramide))
```

Crie uma função que recebe um caracter como parâmetro e conta quantas vezes ele está presente no arquivo "palavras.txt"

```python
def contar_caractere():
    marcus = input('Digite um caractere:')
    
    contador = 0

    with open('palavras.txt', 'r', encoding='utf-8') as texto:
        for linha in texto:
            for letra in linha:
                if letra == marcus:
                    contador += 1
        return contador

contar_caractere()
```
