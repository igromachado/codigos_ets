```python
print("=== Calculadora ===")
print("1 - Soma.\n2 - Subtração.\n3 - Divisão.\n4 - Multiplicação.\n5 - Sair.")

while True:
    try:
        
        a = int(input("Digite o valor de A:"))
        b = int(input("Digite o valor de B:"))
        c = int(input("Qual operação deseja fazer?"))
        
        match c:
            case 1:
                print(f"Resultado da operação: {a + b}")
            case 2:
                print(f"Resultado da operação: {a - b}")
            case 3:
                print(f"Resultado da operação: {a / b}")
            case 4: 
                print(f"Resultado da operação: {a * b}")
            case 5:
                print("Saindo do sistema...")
                break
            case _:
                print("Opção inválida!")
    except ValueError:
        print("Entrada inválida.")  
        
    except ZeroDivisionError:    
        print("Não é possível dividir por 0.")    
```

Exercício 2

```python
original = []

print("=== Verificador de palíndromos ===")

while True:
    try:
        ori = input("Digite uma palavra ou frase:").lower()
        
        for caractere in ori:
            if caractere != " ":
                original.append(caractere)
                tratado = "".join(original)
                
                if tratado.isalpha() == False:
                    raise ValueError("Digite apenas caracteres do alfabeto.")
        
        original.reverse()
        invertida = "".join(original)
        
        if tratado == invertida:
            print(f"{ori.capitalize()} é um palíndromo.")
        else:
            print(f"{ori.capitalize()} não é um palíndromo.")

    except ValueError as erro:
        print(f"Erro: {erro}")
        
    finally:
        print("Encerrando sistema...")
        break
```
