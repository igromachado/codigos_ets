exercício 1 - Faça um programa que pede ao usuário 10 números e armazene-os em uma lista. Depois, peça outro número para o usuário e mostre quantos dos 10 números da lista são inferiores ao valor digitado.

```python
outra_lista = []
lista = []

for i in range(10):
    absurdo = int(input("Digite 10 números: "))
    lista.append(absurdo)

esse_cara = int(input("Digite o número para relacionar: "))

print(f"Lista original: {lista}")

soma = 0

for i in lista:
    if esse_cara > i:
        outra_lista.append(i)
        soma = soma + 1

print(f"{soma} números são menores que o valor digitado: {outra_lista}")
```

exercício 2 - Crie um programa que fatore um número escolhido pelo usuário.

```python
numero = int(input("Digite um número para fatoriar: "))
marcus = numero

while numero > 1:
    numero -= 1
    marcus = marcus * numero

print(marcus)
```

exercício 3 - Cálculo de uma quantidade X de números de sequência de Fibonacci, sendo X um valor inserido pelo usuário.

```python
numero = int(input("Digite a quantidade X da sequência de Fibonacci : "))

lista = []
numero1 = 0
numero2 = 1
resultado = 0

while numero > 0:

    resultado = numero1 + numero2
    lista.append(str(resultado))

    numero1 = numero2
    numero2 = resultado

    numero-=1

resultado = ", ".join(lista)
print(f"A sequência de Fibonacci é: {resultado}")
```

exercício 4 - Com o uso de um dicionário, crie o cadastro de X alunos e então cadastre o nome e o hobby de cada um. Após o cadastro de todos, mostre no terminal todos os alunso cadastrados.

```python
dici = {}

print("=========== CADASTRO DE ALUNOS ===========")

print("Quantos alunos deseja cadastrar? \n")
quantidade = int(input("Quantos alunos deseja cadastrar? "))

for i in range(quantidade):
    print(f"Cadastro do {i+1}º aluno: ")
    nome = input("Digite o nome do aluno: ").capitalize()

    hobby = input("Digite o hobby do aluno: ")
    dici[nome] =  hobby

    print(f"Aluno {nome} cadastrado com sucesso!\n")

print("=========== ALUNOS CADASTRADOS ===========")

for chave,valor in dici.items():
    print(f"Nome: {chave}\nHobby: {valor}\n")
```

exercício 5 - Dados um número X inserido pelo usuário, mostre todos os números primos inferiores a X.

```python
lista = []       

def e_primo(x):
    primo = True
    for i in range(x-1,1,-1):
        if x % i == 0:
            primo = False
            break
    return primo

op = int(input("Digite um número: "))

lista = []

for i in range(op+1):
    if e_primo(i):
        lista.append(i)

print(lista)
```

a) faça o exercício utilizando while + condição.
b) faça o exercício utilizando while True.

```python
import random as r

sorteio = r.randint(0,11)

while True:

    palpite = int(input("Digite seu palpite: "))

    if palpite == sorteio:
        print("Você acertou o número secreto!")
        break
    
    else:
        print("Palpite incorreto... Tente novamente!")
```

exercício 2 - faça um jogo de adivinhação onde o programa irá gerar um número aleatório e o usuário precisa acertar o número. a cada palpite o programa incremente um contador, e avisa se o palpite é maior ou menor que o número escolhido. quando acertar aparece uma mensagem avisando que você acertou e quantos palpites foram necessários.

```python
import random as r

sorteio = r.randint(0,10)

tentativas = 1

while True:

    palpite = int(input("Digite seu palpite: "))

    if palpite == sorteio:
        print(f"Você acertou o número secreto em {tentativas} tentativas!")
        break
    
    elif palpite < sorteio:
        print("Chuta mais alto!")
        tentativas += 1

    elif palpite > sorteio:
        print("Chuta mais baixo!")
        tentativas += 1
```
