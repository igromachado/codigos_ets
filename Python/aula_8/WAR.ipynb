WAR

```python
import random as r

def war (atq,dfs):

    lista_atq = []
    lista_dfs = []

    atq = 50
    dfs = 50

    pontos_atq = 0
    pontos_dfs = 0
    
    for i in range(6):
        lista_atq.append(r.randint(1,6))
        lista_dfs.append(r.randint(1,6))
    
    print("A - D")

    for i in range(6):

        print(f"{lista_atq[i]} - {lista_dfs[i]}")

        if lista_atq[i] > lista_dfs[i]:
            dfs -= 1
            pontos_atq += 1

        elif lista_dfs[i] > lista_atq[i]:
            pontos_dfs += 1
            atq += 1

        elif lista_atq[i] == lista_dfs[i]:
            pontos_dfs += 1
            atq += 1
    print(f"Ataque: {pontos_atq}\nDefesa: {pontos_dfs}")

war(1,1)
```
