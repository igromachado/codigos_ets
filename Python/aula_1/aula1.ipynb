#### EXERCÍCIO 2 - Printar os times conforme suas divisões, formatando os outputs.

```python
lista_times = ['1_palmeiras','2_coritiba','1_corintians','3_juventude','2_fluminense','3_bahia',
               '1_cuiaba', '2_cascavel', '3_ponte preta','2_parana clube', '3_volta redonda']

primeira = []
segunda = []
terceira = []

for item in lista_times:
    
    if item.startswith('1'):
        primeira.append(item[2:].capitalize())
    if item.startswith('2'):
        segunda.append(item[2:].capitalize())
    if item.startswith('3'):
        terceira.append(item[2:].capitalize())

    #o [2:] pega as letras a partir do índice 2 até o fim

print(f"Primeira divisão: {primeira}")
print(f"Segunda divisão: {segunda}")
print(f"Terceira divisão: {terceira}")
    
```

#### CORREÇÃO

```python
lista_times = ['1_palmeiras','2_coritiba','1_corintians','3_juventude','2_fluminense','3_bahia',
               '1_cuiaba', '2_cascavel', '3_ponte preta','2_parana clube', '3_volta redonda']

primeira = []
segunda = []
terceira = []

for item in lista_times:
    
    match item[0]:
        case '1':
            tratado = item.split("_")
            primeira.append(tratado[1].title())
    match item[0]:
        case '2':
            tratado = item.split("_")
            segunda.append(tratado[1].title())
    match item[0]:
        case '3':
            tratado = item.split("_")
            terceira.append(tratado[1].title())
            
print(f"Primeira divisão: {primeira}")
print(f"Segunda divisão: {segunda}")
print(f"Terceira divisão: {terceira}")
```

#### EXERCÍCIO 3 - Cadastro de usuários

```python

nomes_cadastro = []
emails_cadastro = []
contador = 0
continuar = True

print("=== Sistema de Cadastro de Usuários ===\n")

while continuar != False:

#cadastro do nome
    nome = input("Digite o nome completo do usuário:")

    nome_tratado = nome.lower()
    nome_tratado = ' '.join(nome_tratado.split())
    nome_tratado = nome_tratado.title()

    print(f"Digite o nome completo: {nome}")
    nomes_cadastro.append(nome_tratado)

#cadastro do e-mail
    email = input("Digite o e-mail do usuário:")
    
    email_tratado = email.lower()
    email_tratado = ' '.join(email_tratado.split())

#verificação do e-mail      
    if not email.endswith('gmail.com') and not email.endswith('hotmail.com') and not email.endswith('outlook.com') and not email.endswith('yahoo.com'):
        print(f"Atenção: O e-mail {email} possui um domínio desconhecido!")
        
    else:
        print(f"Digite o e-mail: {email}")
    
    emails_cadastro.append(email_tratado)
            
#verificação de continuidade    
    continuar = input("Deseja cadastrar outro usuário? (S/N): ").lower()
    print(f"Deseja cadastrar outro usuário? (S/N): {continuar}")
    
    if continuar == 'n':
        continuar = False
    elif continuar == 's':
        continuar = True
    else:
        print("Opção inválida!")
    
    contador += 1
    
    print()

#mostrar usuários cadastrados
print("=== Relatório de Cadastros ===")

i = 0
for item in range(len(nomes_cadastro)):
    print(f"{nomes_cadastro[i]} | {emails_cadastro[i]}")
    i+=1
print(f"Total de cadastros: {contador}")
```
