2. Faça um programa que gera um número aleatório de 1 a 1000. O usuário deve tentar acertar qual o número foi gerado, a cada tentativa o programa deverá informar se o chute é menor ou maior que o número gerado, caso o número escolhido pelo usuário estiver a distância de 3 para mais ou menos do número sorteado deverá exclusivamente informar que está perto de acertar. O programa acaba quando o usuário acerta o número gerado. O programa deve informar em quantas tentativas o número foi descoberto.

```python
import random

aleatorio = random.randint(1,1000)
tentativas = 1
palpite = 0

while palpite != aleatorio:
    palpite = int(input("Digite seu palpite:"))

    print(f"Tente acertar o número sorteado: {palpite}")
    
    if aleatorio - 3 == palpite or aleatorio - 2 == palpite or aleatorio - 1 == palpite:
        print("Quase!!")
        tentativas = tentativas + 1
    elif aleatorio + 3 == palpite or aleatorio + 2 == palpite or aleatorio + 1 == palpite:
        print("Quase!!")
        tentativas = tentativas + 1
    elif palpite < aleatorio:
        print("Muito baixo.")
        tentativas = tentativas + 1
    elif palpite > aleatorio:
        print("Muito alto.")
        tentativas = tentativas + 1

print(f"Parabéns você acertou! O número é {palpite} \nForam necessárias {tentativas} tentativas.")
```

6. Desenvolva uma função que:
• Solicite ao usuário o número de linhas e colunas de uma matriz.
• Preencha a matriz com valores aleatórios de 0 a 9.
• Exiba a matriz formatada na tela.
• Utilize um for aninhado para percorrer a matriz e calcular a soma dos números pares. Exiba o resultado da soma.

```python
import random

def mostrar_matriz():
    linhas = int(input("Digite o número de linhas da matriz:"))
    print(f"Digite o número de linhas: {linhas}")
    colunas = int(input("Digite o número de colunas da matriz"))
    print(f"Digite o número de colunas {colunas}")
    
    matriz = []
    
    for i in range(linhas):
        linha = []
        for j in range(colunas):
            linha.append(random.randint(0,9))
        matriz.append(linha)
        
    print("\nMatriz gerada:")
    
    for linha in matriz:
        for valor in linha:
            print(f"{valor:3}",end="")
        print()
    
    soma = 0
    
    for i in range(linhas):
        for j in range(colunas):
            if matriz[i][j] % 2 == 0:
                soma = soma + matriz[i][j]
    print(f"Soma dos pares: {soma}")

mostrar_matriz()   
```

7. A Tabela ASCII (American Standard Code for Information Interchange) é um padrão de codificação de caracteres utilizado em computadores e dispositivos eletrônicos. Ela define números inteiros que representam caracteres como letras, números, símbolos e comandos de controle. A tabela ASCII contém 256 caracteres, divididos em duas partes:

• ASCII Padrão (0-127) → Contém caracteres básicos, incluindo: caracteres de controle (0-31) → Exemplo: \n (nova linha), \t (tabulação), números (48-57) → 0-9, letras maiúsculas (65-90) → A-Z, letras minúsculas (97-122) → a-z, símbolos e pontuação (32-47, 58-64, etc.) → @, #, $, %, &, *, etc.

• ASCII Estendido (128-255) → Contém caracteres especiais, como acentos (á, é, ç) e símbolos gráficos.
Nesse primeiro momento vamos analisar apenas a tabela ASCII Padrão com 128 caracteres conforme abaixo.

No Python, as funções ord() e chr() são usadas para trabalhar com a Tabela ASCII. Elas permitem converter entre caracteres e seus códigos numéricos.
•ord() - Obtém o código ASCII de um caractere. Por exemplo, ord(‘W’) = 87.
•chr() - Obtém o caractere a partir do código ASCII. Por exemplo, chr(87) = ‘W’.

Dessa forma, analise o código abaixo e explique seu funcionamento.

```python
p = input('Digite p: ')
n = int(input('Digite n: '))

def funcao(p,n):
    lista = []
    for c in p:
        f = chr((ord(c) - ord('a') + n) % 26 + ord ('a'))
        lista.append(f)
    print(''.join(lista))
    
funcao(p,n)
```
