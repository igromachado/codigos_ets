```python
import random as r

LINHAS = 5
COLUNAS = 5
ELEMENTOS = 3
maior = 0

lista_lista_lista = []

for linhas in range(LINHAS):
    linha = []
    
    for colunas in range(COLUNAS):
        elemento = []
        
        for elementos in range(ELEMENTOS):
            numero = r.randint(0,25)
            elemento.append(numero)
            
        for maiores in range(ELEMENTOS):
            if maiores > maior:
                maior = maiores
                indice = elemento.index(maior)
                
        linha.append(indice)
    lista_lista_lista.append(linha)
                
    print(linha)
```
