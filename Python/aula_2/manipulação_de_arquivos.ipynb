```python
#Sobreescreve o arquivo por completo
with open('./anotacoes.txt', 'w') as anotacoes: 
    anotacoes.writelines(['Sobre\n', 'escrevendo\n'])

#Adiciona ao final
with open('./anotacoes.txt', 'a') as anotacoes:
    anotacoes.writelines(['Adicio\n', 'nando\n'])

#Lê o arquivo 
with open('./anotacoes.txt', 'r') as anotacoes:
    anotacoes.readline()
    for line in anotacoes.readlines():
        print(line)
```
