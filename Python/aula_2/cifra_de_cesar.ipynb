#### Cifra de cesar - codificar palavra

```python
p = input("Digite a palavra que deseja codificar:").lower()
lista = []

for caractere in p:
    codigos = ord(caractere) + 3
    
    if codigos > 122:
        codigos = codigos - 26
        lista.append(chr(codigos))
    
    else:
        lista.append(chr(codigos))
        
resultado = "".join(lista)

print(f"Palavra: {p} \nPalavra codificada: {" ".join(resultado.split('#'))}")
```

#### Cifra de cesar - descodificar palavra

```python
p = input("Digite a palavra que deseja descodificar:").lower()
lista = []

for caractere in p:
    codigos = ord(caractere) - 3
    
    if codigos < 97:
        codigos = codigos + 26
        lista.append(chr(codigos))
    
    else:
        lista.append(chr(codigos))
        
resultado = "".join(lista)

print(f"Palavra codificada: {p} \nPalavra descodificada: {" ".join(resultado.split('7'))}")
```
