```python
print("=== Alunos Cadastrados ===\nNOME\tE-MAIL\tSENHA")

with open('./alunos.txt', 'r') as alunos:
    alunos.readline()
    for line in alunos.readlines():
        separado = line.split(";")
        print("\t".join(separado))

lista = []

print("\n=== Cadastro de novo Aluno ===")

#inputs do usuário
nome = input('Digite o nome do aluno:').title()
print(f"Nome do aluno: {nome}")

email = input('Digite o e-mail do aluno:').lower()

if not email.endswith('gmail.com') and not email.endswith('hotmail.com') and not email.endswith('outlook.com') and not email.endswith('yahoo.com'):
        print(f"Atenção: O e-mail {email} possui um domínio desconhecido!")
        
print(f"E-mail do aluno: {email}")

senha = input('Digite a senha do aluno (somente letras minúsculas):').lower()

#codificar a senha
cript = ""
for letra in senha:
    cript += chr(((ord(letra) + 3 - ord('a')) % 26 + ord('a')))
    
print(f"Senha do aluno: {cript}")

#adicionar na lista
lista.append(nome)
lista.append(email)
lista.append(cript)

#deixar o resultado no formato pedido
resultado = ";".join(lista)

print("Aluno cadastrado com sucesso!")

with open('./alunos.txt', 'a') as alunos:
    alunos.write(f"\n{resultado}")
```
