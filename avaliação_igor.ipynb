```python
df = pd.read_csv('vendas.csv')

print(df['data'])

# def formatar_data(linha):
#     data = linha['data']
#     data = data.replace('/','-')
#     data = pd.to_datetime(linha['data'], errors = 'coerce')

# df['data'] = df.apply(formatar_data,axis=1)

# print(df['data']

df['data'] = pd.to_datetime(df['data'], errors='coerce')

print(df['data'])

print(df['data'].isna().sum())
```

```python
df['desconto'] = np.where(df['desconto'].isnull(), 0,df['desconto'])

print(df.isnull().sum())
```

```python
media = round(df['preco'].mean(),2)
print(media)

df['preco'] = np.where(df['preco'].isnull(), media,df['preco'])

df.head()
```

```python
df = df[df['quantidade'] > 0]
df.head()
```

```python
filtrado = df[['produto','categoria','preco','quantidade','desconto']]
filtrado.head(10)
```

```python
filtrado.tail(5)
```

```python
df['valor_bruto'] = round(df['preco'] * df['quantidade'],2)
df['valor_liquido'] = round(df['valor_bruto'] * (1 - df['desconto']),2)
df['ticket_medio'] = round(df['valor_liquido'] / df['quantidade'],2)

df.head()
```

```python
filtrado = df[(df['categoria'] == 'Eletrônicos') & (df['valor_liquido'] > 1000)]
filtrado = filtrado[['produto','categoria','preco','quantidade','desconto']]
```

```python
faturamento_total = round(df['valor_liquido'].sum(),2)

print(f"Faturamento total: R${faturamento_total}")

print(f"Média de faturamento por venda: R${round((faturamento_total / df.size),2)}")

print(f"Quantidade total de itens vendidos: {df.shape[0]}")

```

```python
quantidade_vendida = df.groupby('produto')['quantidade'].sum()
quantidade_vendida = quantidade_vendida.sort_values(ascending=False)

print(f"Produto mais vendido: {quantidade_vendida[0]}")
```

```python
maior_faturamento = df.groupby('produto')['valor_bruto'].sum()

maior_faturamento = maior_faturamento.sort_values(ascending=False)

print(f"Produto com maior faturamento: {round(maior_faturamento[0],2)}")
```

```python
df = pd.read_csv('vendas.csv').dropna()

df['valor_total'] = df['preco'] * df['quantidade']
```

```python
contagem = df.groupby('produto')['quantidade'].sum()

plt.bar(contagem.index,contagem)
plt.xticks(rotation=45)
plt.title('Quantidade de Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Quantidade')

plt.show()
```

```python
pizza = df.groupby('categoria')['categoria'].count()

plt.pie(pizza,labels=pizza.index,autopct="%1.1f%%")
plt.title('Participação das Categorias no Faturamento Total')
plt.legend()
plt.show()
```

```python
print("Escolha o efeito: \n1 - BGR para Luv\n2 - BGR para HSV\n3 - BGR para RGB.")
escolha = input("Digite a opção (1/2/3): ")

cap = cv2.VideoCapture('flor.mp4')

fourcc = cv2.VideoWriter_fourcc(*'XVID')
output = cv2.VideoWriter('novo_video.mp4', fourcc, 30, (1280,720))

match escolha:
    
    case "1":
        cor = cv2.COLOR_BGR2Luv
    case "2":
        cor = cv2.COLOR_BGR2HSV
    case "3":
        cor = cv2.COLOR_BGR2RGB
    case _:
        print("Comando incorreto.")

while (cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        frame = cv2.cvtColor(frame, cor)
        frame2 = cv2.resize(frame,(1280,720))
        output.write(frame2)
    else:
        break
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cap.release()
output.release()
```
