```python
import cv2
import numpy as np

def function(event,x,y,flag,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),10,(blue,red,green),-1)

red = 0
blue = 0
green = 0
font = cv2.FONT_HERSHEY_SIMPLEX

img = np.zeros((1080,1920,3), np.uint8)
cv2.namedWindow('Desafio 2')
cv2.setMouseCallback('Desafio 2', function)

while True:

    cv2.putText(img,f'Red = {red}',(50,100),font, 1,(255,255,255),4,cv2.LINE_AA)
    cv2.putText(img,f'Blue = {blue} ',(50,150),font, 1,(255,255,255),4,cv2.LINE_AA)
    cv2.putText(img,f'Green = {green}',(50,200),font, 1,(255,255,255),4,cv2.LINE_AA)
    cv2.imshow('Desafio 2',img)
    
    t = cv2.waitKey(20) & 0xff
    
    if t == ord('r'):
        red += 10
        if red > 255:
            red = 0
    
    if t == ord('b'):
        blue += 10
        if blue > 255:
            blue = 0
            
    if t == ord('g'):
        green += 10
        if green > 255:
            green = 0
    
    if t == ord('x'):
        break
    
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture('video_car.mp4')

fourcc = cv2.VideoWriter_fourcc(*'XVID')
output = cv2.VideoWriter('novo_video.mp4', fourcc, 20.0, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret == True:
        frame = cv2.flip(frame, 0)
        output.write(frame)
        
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
        
cap.release()
output.release()
cv2.destroyAllWindows()
```

```python
import cv2

video = cv2.VideoCapture('video_car.mp4')

fourcc = cv2.VideoWriter_fourcc(*'XVID')
output = cv2.VideoWriter('carro_psicodélico.mp4', fourcc, 20, (300,300))

while True:
    ret, frame = video.read()

    if ret == True:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        frame2 = cv2.resize(frame,(300,300))
        output.write(frame2)
        
        cv2.imshow('frame',frame2)
        if cv2.waitKey(33) & 0xff == ord('q'):
            break
    else:
        break
    
video.release()
output.release()
cv2.destroyAllWindows()
```
