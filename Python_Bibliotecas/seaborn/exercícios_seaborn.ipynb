```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CSVs/flights.csv')

filtro = df['year'] == 1950
df = df[filtro]

sns.catplot(x='passengers',y='month',data=df, palette='deep',kind='bar')

plt.show()
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

sns.histplot(data=df, x='total_bill', hue='sex', kde=True)
plt.show()

#Homens costumam gastar mais que as mulheres, tendo um pico de gastos em 15 (total_bill).
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

sns.catplot(data=df,x='day', y='total_bill', kind='box',hue='smoker')
plt.show()

#Sábado é o dia com mais outliers; Fumantes possuem uma maior concentração de dados em valores maiores, por conta disso, fumantes gastam mais
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

sns.relplot(data=df,x='total_bill',y='tip',hue='day',size='servings')

plt.show()

#Esse fator pode influenciar na quantidade de gorgeta, porém mesas com mais pessoas não signicam necessariamente que darão uma gorgeta maior.
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

df.drop(['sex','smoker','day','time'],axis=1, inplace=True)

sns.heatmap(df.corr(), annot=True)

plt.show()

#O valor total da conta (total_bill) influencia mais no valor da gorgeta.
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

g = sns.FacetGrid(data=df, col='time')

g.map_dataframe(sns.histplot, x='total_bill')

plt.show()

#Sim, nos jantares há uma maior ocorrência em total_bill, o que indica uma quantidade maior de atendimentos.
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

sns.violinplot(data=df,x='day',y='tip',hue='sex',split=True)
plt.show()

#Sábado é o dia com maior variação nas gorjetas, isso é indicado pela altura que o gráfico apresenta.
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

sns.pairplot(df,hue='sex')
plt.show()

#Os gráficos apontam que na maioria dos dias são as pessoas do sexo masculino que gastam mais.
```

```python
df = pd.read_csv('CSVs/tips.csv',sep=';')

sns.catplot(data=df,x='day',y='total_bill',hue='sex',kind='bar')
plt.show()
```
