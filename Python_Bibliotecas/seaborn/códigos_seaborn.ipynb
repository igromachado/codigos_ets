```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('CSVs/tips.csv',sep=';')
```

```python
sns.relplot(x='total_bill', y='tip', data = df)
```

```python
sns.relplot(x='total_bill', y='tip',hue='day', data = df)
```

```python
sns.relplot(x='total_bill', y='tip', style='smoker', data = df)
```

```python
sns.relplot(x='total_bill', y='tip', size='servings', data = df)
```

```python
sns.relplot(x='total_bill', y='tip', hue='day',style= 'smoker', size='servings', data = df)
```

```python
sns.relplot(x='total_bill',y='tip',col='time',
            hue='time',size='servings',style='sex',
            palette=['b','r'], sizes=(10,100),data=df)
```
