```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('CSVs/fish_detail.csv')
df.columns
```

```python
df.drop(['Used In','Description'],axis=1,inplace=True)
```

```python
def get_difficulty(line):
    return line['Difficulty & Behavior'][:2]

df['Difficulty'] = df.apply(get_difficulty,axis=1)
df.head()
```

```python
def get_behavior(line):
    return line['Difficulty & Behavior'][3:]

df['Behavior'] = df.apply(get_behavior,axis=1)
df.head()
```

```python
df.drop('Difficulty & Behavior',axis=1,inplace=True)
```

```python
df.head()
```

```python
dft = df.copy()
dft['Location'] = dft['Location'].str.replace('\n', ',')
dft['Location'] = dft['Location'].str.split(',')
dft = dft.explode('Location')
dft.head()
```

```python
dfn = df.copy()
dfn['Season'] = dfn['Season'].str.replace('\n',',')
dfn['Season'] = dfn['Season'].str.replace(',(All Seasons on Ginger Island)','',regex=False).str.strip()
dfn['Season'] = dfn['Season'].str.split(',')
dfn = dfn.explode('Season')
dfn
```

```python
dff = df.copy()

dff['Location'] = dff['Location'].str.replace('\n', ',', regex=False)
dff['Location'] = dff['Location'].str.replace('*Requires fishing level 4*','',regex=False).str.strip()
dff['Location'] = dff['Location'].str.split(',')

dff['Season'] = dff['Season'].str.replace('\n',',')
dff['Season'] = dff['Season'].str.replace(',(All Seasons on Ginger Island)','',regex=False).str.strip()
dff['Season'] = dff['Season'].str.split(',')

dff = dff.explode('Location')
dff = dff.explode('Season')

dff = dff.dropna().reset_index(drop=True)

dff
```
