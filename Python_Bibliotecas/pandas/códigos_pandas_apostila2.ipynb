```python
def calculate_percentage(val, total):
#Calculates the percentage of a value over a total
    percent = float(val / total)
    beautiful_percent = ("%.2f" % (percent * 100) + "%")
    return beautiful_percent

total = titanic.Survived.count()
nao_sobreviveu = titanic.Survived[titanic["Survived"] == 0].count()
sobreviveu = titanic.Survived[titanic["Survived"] == 1].count()
print(total)
print(nao_sobreviveu)
print(sobreviveu)
```

```python
total = titanic.Survived.count()

mortos_com_familia = titanic.Survived[(titanic["Survived"] == 0) & (titanic["Relatives"] > 0)].count()
mortos_sem_familia = titanic.Survived[(titanic["Survived"] == 0) & (titanic["Relatives"] == 0)].count()
print("Mortos com família:", mortos_com_familia,"\nMortos sem família:", mortos_sem_familia)

homens = titanic.Sex[titanic["Sex"] == "male"].count()
homens_mortos = titanic.Sex[(titanic["Sex"] == "male") & (titanic["Survived"] == 0)].count()
print("Homens:", homens,"| Homens mortos:", homens_mortos)

mulheres = titanic.Sex[titanic["Sex"] == "female"].count()
mulheres_mortas = titanic.Sex[(titanic["Sex"] == "female") & (titanic["Survived"] == 0)].count()
print("Mulheres:", mulheres,"| Mulheres mortas:", mulheres_mortas)

crianças = titanic.Survived[titanic["Age Group"] == "Criança"].count()
print("Porcentagem de crianças:",round((crianças/total)*100,2))

adultos = titanic.Survived[titanic["Age Group"] == "Adulto"].count()
print("Porcentagem de adultos:",round((adultos/total)*100,2))

adolescente = titanic.Survived[titanic["Age Group"] == "Adolescente"].count()
print("Porcentagem de adolescentes:",round((adolescente/total)*100,2))

idosos = titanic.Survived[titanic["Age Group"] == "Idoso"].count()
print("Porcentagem de idosos:",round((idosos/total)*100,2))

primeira = titanic.Survived[titanic["Pclass"] == 1].count()
segunda = titanic.Survived[titanic["Pclass"] == 2].count()
terceira = titanic.Survived[titanic["Pclass"] == 3].count()
print("1ª classe:", primeira,"2ª classe:", segunda,"3ª classe:", terceira)

p_morto = titanic.Survived[(titanic["Pclass"] == 1) & (titanic["Survived"] == 0)].count()
s_morto = titanic.Survived[(titanic["Pclass"] == 2) & (titanic["Survived"] == 0)].count()
t_morto = titanic.Survived[(titanic["Pclass"] == 3) & (titanic["Survived"] == 0)].count()
print("mortos da 1ª classe:", p_morto, "| mortos da 2ª classe:",s_morto, "| mortos da 3ª classe:",t_morto)

```

```python
agrupado = titanic.groupby("Age Group").count()

def porcentagem(linha):
    return (f"{round((linha["PassengerId"] / agrupado["PassengerId"].sum()) * 100,2)}%")
    
per = agrupado.apply(porcentagem, axis = 1)

agrupado["Percentual"] = per

print(agrupado[["PassengerId","Percentual"]])
```
