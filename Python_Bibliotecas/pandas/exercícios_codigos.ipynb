```python
tips = pd.read_csv("bases/tips.csv", sep=";")

tips.head()

def gorgeta(linha):
    return (round((linha['tip'] / linha['total_bill']) * 100,2))

per = tips.apply(gorgeta,axis=1)

tips["tip_pct"] = per

tips[tips["tip_pct"] == tips["tip_pct"].max()]
```

```python
mediaporsexo = round(tips.groupby("sex")['tip'].mean(),2)
print(mediaporsexo)
```

```python
print(tips['servings'].count())
```

```python
media = tips['tip'].mean()

def pagaruim(linha):
    if linha['tip'] < media:
        return "mau pagador"
    else:
        return "bom pagador"
    
x = tips.apply(pagaruim,axis=1)

tips["class"] = x

tips.head()
```

```python
maus = tips[tips['class'] == 'mau pagador'].groupby("day")['class'].count()

maus[maus == maus.max()]
```

```python
maus = tips[tips['class'] == 'mau pagador']

fumantes = maus.groupby('smoker')['class'].count().to_frame(name="Maus pagadores")
print(fumantes)
```
