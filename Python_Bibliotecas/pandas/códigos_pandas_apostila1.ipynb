```python
import pandas as pd
titanic = pd.read_csv("bases/titanic.csv", sep=",")
def soma_sibsp_parch(linha):
    return linha["SibSp"] + linha["Parch"] # ou seja, todos os parentes dentro do navio

nova_coluna = titanic.apply(soma_sibsp_parch, axis=1)

nova_coluna.head()
```

```python
def verifica_idade(linha):
    if linha["Age"] < 12:
        return "Criança"
    
    elif linha["Age"] < 18:
        return "Adolescente"
    
    elif linha["Age"] < 65:
        return "Adulto"
    
    elif linha["Age"] >= 65:
        return "Idoso"
    else:
        return "nulo"

faixa_etaria = titanic.apply(verifica_idade, axis=1)

titanic["Age Group"] = faixa_etaria

print(titanic[["Age","Age Group"]].head(50))
```

```python
titanic["Relatives"] = nova_coluna
titanic.to_csv("bases/titanic_alterado.csv",index=False)
```
