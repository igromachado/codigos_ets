```python
import pandas as pd

#mostrar somente as colunas 'Manufacturer','Make','Price','MPG.city','Type','Passengers'
carros = pd.read_csv('bases/Cars93_miss.csv',sep = ",").loc[:,['Manufacturer','Make','Price','MPG.city','Type','Passengers']].dropna()

#deve ser um carro com 5 lugares
carros = carros[carros["Passengers"]== 5] 

#selecione os 10 carros com maior MPG na cidade
carros = carros.sort_values("MPG.city",ascending=False).head(10)

#dos 10 mais econômicos, mostre os 5 modelos mais baratos
carros = carros.sort_values("Price").head(5)
print(carros)
```
