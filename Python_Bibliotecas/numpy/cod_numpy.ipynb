```python
def faz_matriz(n):
    matriz = np.full((n,n),0)

    for i in range(n):
            matriz[i,i] = 1
            matriz[i,n-1-i]+=2
    return matriz

n = 6
print(faz_matriz(n))
```

```python
n = 4

m = np.floor(10*np.random.random((n,n)))
print(m,'\n')

for i in range(n):
    for j in range(n):
        
        if i == 0 and j == 0 and i < n-1 and j < n-1:
            v = m[i+1,j] + m[i,j+1]
            
        if i == 0 and j != 0 and i < n-1 and j < n-1:    
            v = m[i+1,j] + m[i,j+1] + m[i,j-1]
        
        if i != 0 and j == 0 and i < n-1 and j < n-1:
            v = m[i+1,j] + m[i,j+1] + m[i-1,j]
            
        if i != 0 and j != 0 and i < n-1 and j < n-1:
            v = m[i+1,j] + m[i,j+1] + m[i-1,j] + m[i,j-1]      
              
        print(v,end="  ")
    print()
    
```

```python
import numpy as np

n = 4

m = np.random.randint(0,10, (n,n))
print(m,'\n')

for i in range(n):
    for j in range(n):
        
        v = 0
        
        if i+1 < n:
            v += m[i+1,j]
            
        if i-1 >= 0:
            v += m[i-1,j]
        
        if j+1 < n:
            v += m[i,j+1]
        
        if j-1 >= 0:
            v += m[i,j-1]
            
        print(v,end="\t")
    print()
```

```python
import numpy as np

n = 4

m = np.random.randint(0,10, (n,n))
print(m,'\n')

somavizinhos = np.empty_like(m)

for i in range(n):
    for j in range(n):
        
        v = 0
        
        if i+1 < n:
            v += m[i+1,j]
            
        if i-1 >= 0:
            v += m[i-1,j]
        
        if j+1 < n:
            v += m[i,j+1]
        
        if j-1 >= 0:
            v += m[i,j-1]
            
        somavizinhos[i,j] = v
            
print(somavizinhos)
```

```python
n = 4

m = np.random.randint(0,20,(n,n))
mascara = np.empty_like(m)

print(m,"\n")

for i in range(n):
    for j in range(n):
        
        v = 1
        
        if i+1 < n and m[i+1,j] > m[i,j]:
            v = 0
            
        if i-1 < n and m[i-1,j] > m[i,j]:
            v = 0
            
        if j+1 < n and m[i,j+1] > m[i,j]:
            v = 0
            
        if j-1 < n and m[i,j-1] > m[i,j]:
            v = 0

        mascara[i,j] = v
            
print(mascara)
```

```python
def faz_matriz(n):

    m = np.random.randint(0,100,(n,n))
    print(m,'\n')
    
    print(f"Máximo: {m.max()}\nMínimo: {m.min()}\nMédia: {round(m.mean(),2)}\n")

    menos_um = m<m.mean() #menores que a média
    zero = m == m.mean() #igual à média
    um = m>m.mean() #maiores que a média
    nove = m==m.max() #igual ao valor máximo
    menos_nove = m==m.min() #igual ao valor mínimo

    m[menos_um] = -1
    m[zero] = 0
    m[um] = 1
    m[nove] = 9
    m[menos_nove] = -9
     
    print(m,'\n')

valor = 4

faz_matriz(valor)
```
