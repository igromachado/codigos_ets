```python
import numpy as np

notas = np.random.randint(0,11,(100,6))

print(f'Shape: {notas.shape}')
print(f'Dimensões: {notas.ndim}')
print(f'Tamanho: {notas.size}')
```

```python
douzi = notas.reshape(50,12)
antigo = douzi.reshape(100,6)

print(douzi)
print()
print(antigo)
```

```python
print(notas[10,3])
print()
print(notas[20:30,2:4])
```

```python
print(notas[1])

notas[notas < 5] = 5

print(notas[1])
```

```python
print(notas[1])

notas = notas + 0.5

print(notas[1])
```

```python
media = round(notas.mean())

print(media)

print(notas[1])

diff = notas - media

print(diff[1])
```

```python
print(notas.mean(axis=0)) #percorre colunas
print(notas.mean(axis=1)) #percorre linhas 
```

```python
print(notas.max(axis=1))
print(notas.std()) #desvio padrão
```

```python
maior = notas.max()

indices = np.argwhere(notas == maior)

print(indices)
```

```python
print(np.where(notas >= 7,True,False))
```

```python
sete_maior = notas[notas >= 7]
print(sete_maior)
```

```python
medias = notas.mean(axis=1)

medias = medias[medias>=7]

print(medias.size)
```

```python
grupo_a = np.array([0,1,2,3,4])
grupo_b = np.array([3,4,5,6,7])

#inteseção
print(np.intersect1d(grupo_a,grupo_b))

#união
print(np.union1d(grupo_a,grupo_b))

#diferença
diferença = grupo_a - grupo_b
print(diferença)
```

```python
x_max = notas.max()
x_min = notas.min()

normalizadas = (notas - x_min) / (x_max - x_min)

print(normalizadas)
```
