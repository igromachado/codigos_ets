```python
import time
```

```python
print("Digite a frase abaixo o mais rápido possível: \n")
frase = 'The quick brown fox jumps over the lazy dog'

char = 0
for i in frase:
    char += 1

print(frase)
print("\nPressione ENTER para começar: \n")
input()

for i in range(2,0,-1):
    print(i)
    time.sleep(1)
    
start = time.time()
resposta = input('Digite aqui: ')
end = time.time()

total = round(end - start,2)
velocidade = round(char / total,2)

print(f"\nRESULTADO\nTempo total: {total} \nVelocidade: {velocidade} caracteres por segundo")

if resposta == frase:
    print("Você digitou corretamente!")
else:
    print("Você errou algum caractere...")
```
