```python
def arrumar_arquivos():
    
    path = os.path.abspath('arquivos')

    if os.path.isdir('arquivos/Imagens') == False:
        os.mkdir('arquivos/Imagens')

    if os.path.isdir('arquivos/Documentos') == False:
        os.mkdir('arquivos/Documentos')

    if os.path.isdir('arquivos/CSVs') == False:
        os.mkdir('arquivos/CSVs')

    arquivos = os.listdir('arquivos')

    for i in arquivos:
        
        origem = os.path.join(path, i)
        
        arq = i.split('.')
        
        if len(arq) == 2:
            
            match arq[1]:
                
                case 'jpg' | 'png':
                    novo_local = os.path.join(path, 'Imagens', i)
                
                case 'csv':
                    novo_local = os.path.join(path, 'CSVs', i)

                case 'pdf':
                    novo_local = os.path.join(path, 'Documentos', i)

            os.rename(origem,novo_local)
                    
arrumar_arquivos()
```
