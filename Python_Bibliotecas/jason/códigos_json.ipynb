```python
import json
with open('data.json','r',encoding='utf-8') as file:
    conteudo = file.read()
    
y = json.loads(conteudo)
print(y)
```

```python
for i in y['vendas']:
    print(i['produto'])
```

```python
soma = 0
for i in y['vendas']:
    soma = i['valor'] + soma
    
print(soma)
```

```python
produtos = {}

for i in y['vendas']:
    valor = i['valor']
    produto = i['produto']
    
    if produto in produtos:
        produtos[produto] += valor
        
    else:
        produtos[produto] = valor

mais_vendido = ''
maior = 0
for chave, valor in produtos.items():
    if valor > maior:
        maior = valor
        mais_vendido = chave
        
print(mais_vendido)
    
```

```python
total_vendedor = {}

for i in y['vendas']:
    vendedor = i['vendedor']
    valor = i['valor']
    
    if vendedor in total_vendedor:
        total_vendedor[vendedor] += valor
        
    else:
        total_vendedor[vendedor] = valor


maior = 0
vendedor = ''

for chave, valor in total_vendedor.items():
    if valor > maior:
        maior = valor
        vendedor = chave
    
```

```python
final = {}

final['total_vendas'] = soma
final['produto_mais_vendido'] = mais_vendido
final['melhor_vendedor'] = vendedor

with open('relatório.json','w',encoding='utf-8') as arq:
    json.dump(final,arq)
```
