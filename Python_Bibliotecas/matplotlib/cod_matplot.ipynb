```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('respiradores.csv', sep=",")

x=df.MES
y=df.TOTAL
plt.bar(x,y, 1,align= "center", edgecolor="black" )
plt.title("COMPRA DE RESPIRADORES POR MÊS")
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.show()
```

```python
total_por_estado = df.sum()[1:-1]
total_por_estado
```

```python
plt.bar(x, total_por_estado, color = 'black', zorder=3)
plt.title('COMPRA DE RESPIRADORES POR ESTADO')
plt.xticks(rotation='vertical')
plt.grid(zorder=0)
plt.show()
```

```python
plt.barh(df.MES,df.TOTAL, color = 'purple')
plt.title('MEU DEUS DO CÉU')
plt.show()
```

```python
plt.plot(df.MES,df.PARANA)
plt.xticks(rotation=45)
plt.grid(linestyle='dashed')
plt.title('DEBÍ TIRAR MAS FOTOS')
plt.show()
```

```python
plt.plot(df.MES,df.PARANA,color='black',marker='o',linewidth=3,markersize=10)
plt.xticks(rotation=45)
plt.title('DE CUANDO TE TUVE')
plt.grid(linestyle='dashed')
plt.show()
```
